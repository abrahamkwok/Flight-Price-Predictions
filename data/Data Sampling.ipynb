```
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Define the input and output file paths
input_file_path = '/work/processed_data.csv'
output_file_path = '/work/filtered_data.csv'

# Set the chunk size (adjust based on available memory)
chunk_size = 100000
```

```
# Open the output file for writing
with open(output_file_path, 'w') as output_file:
    # Iterate over the file in chunks
    for i, chunk in enumerate(pd.read_csv(input_file_path, chunksize=chunk_size)):
        # Perform any necessary processing on the chunk
        # Example: Filter rows for only CA airports
        processed_chunk = chunk[(chunk['startingAirport'] == 'SFO') | (chunk['startingAirport'] == 'OAK') | (chunk['startingAirport'] == 'LAX')]

        # For the first chunk, write headers; for subsequent chunks, skip the headers
        if i == 0:
            processed_chunk.to_csv(output_file, index=False)
        else:
            processed_chunk.to_csv(output_file, index=False, header=False)
```
